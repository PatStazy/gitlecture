### Summary Statistics: First X Variable Pertaining to Y

```python
state_voter_pop = df_2016[["state", "candidatevotes"]].groupby(["state"]).sum().rename(columns={'candidatevotes':'totalvotes'})
state_voter_pop.head()
```

I created a groupby that sums candidatevotes per state in order to get a single value of totalvotes per state. This will be used to merge with my first data-frame in the function below

```python
#create function and dictionary to add new column based on "party_simplified"
def comb_value(row_number, values):
    return values[row_number]
party_dictionary = {'REPUBLICAN' : 'REPUBLICAN', 'DEMOCRAT' : 'DEMOCRAT', 'LIBERTARIAN' : 'THIRD PARTY', 'OTHER' : 'THIRD PARTY'}
df_2016["updated_parties"] = df_2016["party_simplified"].apply(comb_value, args = (party_dictionary,))

#Drop unnecessary columns
df_2016 = df_2016.drop(columns=["party_simplified", "totalvotes"])

#set "updated_parties" to only pertaine to Third Party votes
df_third_party = df_2016[df_2016['updated_parties']=='THIRD PARTY']

#Drop "updated_parties" because it's already known this dataset is only for third party data
df_third_party = df_third_party.drop(columns=["updated_parties"])

#groupby state, and merge state_voter_pop data to get proper value for total votes
state_parties = df_third_party[["state", "candidatevotes"]].groupby(["state"]).sum()
voter_data = state_parties.merge(state_voter_pop, left_on="state", right_index=True)
voter_data.head()
```

```python
#get percentage of third party votes

voter_data['third_party_vote_percentage'] = (voter_data["candidatevotes"]/voter_data["totalvotes"])*100
voter_data.head()
```

I manipulated my dataset (df_2016) in order to get a new column that summed all values for third parties (LIBERTARIAN and OTHER) and gave it one value name: THIRD PARTY. I then grouped by state as I want the data pertaining to each state, added the proper total voter population value and created another column that shows the values of the percentage of each states voter population that voted for a third party. I gave this manipulated data-frame the variable name voter_data

These tables above are all my summary statistics for my first X variable: Total Population of Voters Per State, as it pertains to my Y variable.

### Summary Statistics: Second X Variable Pertaining To Y

```python
voter_data_merge = voter_data.drop(columns=["candidatevotes"])
voter_data_merge.head()
```

Similar to my first table shown in Summary Statistics for X1 Variable, I am creating a new dataset based off of voter_data, which I will use to merge into my second data frame called party_winner_data. This will allow me to add the column third_party_vote_percentage to my second dataframe which pertains to the winning party in each state. 

```python
#create function and dictionary to add new column "winning_party" based off "win"
def comb_value(row_number, values):
    return values[row_number]
party_dictionary_winner = {'R' : 'REPUBLICAN', 'D' : 'DEMOCRAT'}
df2["winning_party"] = df2["win"].apply(comb_value, args = (party_dictionary_winner,))

#drop column "win"
df2_1 = df2.drop(columns=["win"])

#changing values in column "state" so they are all capital letters
df2_1["state"] = df2_1["State"]
df2_1["state"] = df2_1["state"].str.upper()

#grouping by "state" and merging voter_data_merge
party_winner_data = df2_1[["state", "winning_party"]].groupby(["state"]).sum()
party_winner_data = party_winner_data.merge(voter_data_merge, left_on="state", right_index=True)
party_winner_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import urllib.request
```

```python
new_data_url = "https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_income"
response = requests.get(new_data_url)
#aquire URL

soup_object = BeautifulSoup(response.content)

data_table = soup_object.find_all("table", "wikitable sortable")[0]

all_values = data_table.find_all('tr')

median_income_df = pd.DataFrame(columns = ['state', 'Y_2016', 'Y_2015', 'Y_2014', 'Y_2013']) # Create an empty dataframe

ix = 0 # Initialise index to zero

for row in all_values[1:]:
    values = row.find_all('td') # Extract all elements with tag <td>
    # Pick only the text part from the <td> tag
    state = values[1].text
    Y_2016 = values[5].text
    Y_2015 = values[6].text
    Y_2014 = values[7].text
    Y_2013 = values[8].text
    median_income_df.loc[ix] = [state, Y_2016, Y_2015, Y_2014, Y_2013] # Store it in the dataframe as a row
    ix += 1

median_income_df = median_income_df.replace(r'\n','', regex=True)
median_income_df["Y_2016"] = median_income_df["Y_2016"].str.replace('\$', '', regex=True)
median_income_df["Y_2015"] = median_income_df["Y_2015"].str.replace('\$', '', regex=True)
median_income_df["Y_2014"] = median_income_df["Y_2014"].str.replace('\$', '', regex=True)
median_income_df["Y_2013"] = median_income_df["Y_2013"].str.replace('\$', '', regex=True)
median_income_df["Y_2016"] = median_income_df["Y_2016"].str.replace(',', '', regex=True)
median_income_df["Y_2015"] = median_income_df["Y_2015"].str.replace(',', '', regex=True)
median_income_df["Y_2014"] = median_income_df["Y_2014"].str.replace(',', '', regex=True)
median_income_df["Y_2013"] = median_income_df["Y_2013"].str.replace(',', '', regex=True)
median_income_df["Y_2016"] = median_income_df["Y_2016"].astype(str)
median_income_df["Y_2015"] = median_income_df["Y_2015"].astype(str)
median_income_df["Y_2014"] = median_income_df["Y_2014"].astype(str)
median_income_df["Y_2013"] = median_income_df["Y_2013"].astype(str)
median_income_df["state"] = median_income_df["state"].str.strip()
median_income_df["Y_2016"] = median_income_df["Y_2016"].str.strip()
median_income_df["Y_2015"] = median_income_df["Y_2015"].str.strip()
median_income_df["Y_2014"] = median_income_df["Y_2014"].str.strip()
median_income_df["Y_2013"] = median_income_df["Y_2013"].str.strip()
median_income_df["Y_2016"] = median_income_df["Y_2016"].astype(int)
median_income_df["Y_2015"] = median_income_df["Y_2015"].astype(int)
median_income_df["Y_2014"] = median_income_df["Y_2014"].astype(int)
median_income_df["Y_2013"] = median_income_df["Y_2013"].astype(int)
#Remove newline
median_income_df.head()
```

#### Measures used for Evaluating my Regressions

First Measure: R-Squared. The R-Squared value is crucial when evaluating a linear regression as it desplays the degree to which my X-Variables explain the variation of my dependent variable. In this case, my third-regression is the one I preffer, it included 6 X-Variables that explains a total of 19.8% of my dependent variables variation. A problem with R-Squared evaluation is that it can only increase or stay stagnent as more variables are added, meaning in the end, many variable that are included in the regression could actually have zero effect on the relationship with the dependent variable. In simpler terms, it acts as a stat-accumulator. This can be solved by taking into account the values of Adjusted R-Squared.

Second Measure: Adjusted R-Squared. The adjusted R-Squared value is necessary as it keeps the R-Squared values in-check. While a reasercher could continue to add X-Variables in order for R-squared to give a better result, Adjusted R-Squared penalizes the reasercher if there are added X-Variables that do not improve the results. Therefore the Adjusted R-Squared value is incredibly important to my reaserch as while the R-Squared value for my third and fourth regression are virtually the same, by analyzing the Adjusted R-Squared value, I can see that the third regression is actually the most precise regression, as it has the highest Adjusted R-Squared value, meaning I don't have any unnecessary variables.  

Third Measure: AIC and BIC values. The ACI method is used for evaluating how well my regression model fits the data it was generated from. "It assesses the number of independent variables used when building the model and how well the model reproduces the data. The Smaller the AIC value, the better the fit. The BIC value measures the trade-off between model fit and complexity of the model" (Bevans, 2021). Again the smaller the value, the better the fit. Although the first two regression models that I ran had lower AIC and BIC values, they allocate a heavy weight to the amount of independent variables involved in the regression. There are many measures I have to take into account while running these regressions and due to a small increase in the AIC value leading up to the third regression, and a greater increase in my BIC value (mostly due to the addition of 4 new independent variables). I have to take into account my R and Adjusted R-Squared values as well, and since the AIC and BIC values don't have a large increase, but the R-Squared and Adjusted R-Squared values have a pretty profound change, I still preferr the third regression model to the other three.

Fourth Measure: P-Values. The P-Value shows weither my coefficient estimates are statistically significant. "It is the probability of obtaining results at least as extreme as the observed results of a hypothesis test, that assumes the null hypothesis is correct" (Beers, Anderson, 2022). "The null hypothesis proposes that no statistical significance exists in a given observation" (Hayes, Andrson, Kvilhaug, 2022). The P-value is created from running the linear regression model that I created, compared with this null-hypothesis which is used to show the most extreme case (Being my independent variables have no correlation whatsoever with my dependent variable). Overall, the P-value shows the percentage probability that each of my independent variables have of not being at all significant. The P-Value is between 0 and 1. A P-Value of 0.05 or less is considered statistically significant, meaing it predicts the true effect the independent variable has with the dependent occurs 95% of the time. However, just because the P-Value is greater than 0.05 doesn't mean it has no significance, it just concludes that there isn't enough evidence to pressume there is an effect.

Fifth Measure: F-Statistics (F-Test). "F-Statistics evaluate the overall significance of the linear regression model. In short, it tests to see if my linear regression model better fits the data than a model containing no X-Variables. The F-Test "compares my specific model to a model with no independent variables." (Frost, Date Unknown). 

Overall, my linear regression model can be shown like this:

Y = B0 + B1x1 + B2x2 + B3x3 + B4x4 + B5x5 + B6x6 + Bkxk + u 

The F-Test tests on all slope parameters (B0, B1, B2, ...). H0 (Model with no independent variables): B1=B2=...Bk=0, H1 (My Model): at least one of the slope parameters don't equal zero. 

An issue with this test is that the F-Test determines whether all independent variables are jointly significant. It is possible that each independent variable is not significant and yet the F-Test says that all independent variables combined are jointly significant.

#### Objective Function

Our objective function for this regression tree will be used to minimize the error of prediction. Error of prediction: (Y-Y(hat)). What is the use of this? If I want to predict the outcome based on my variables for the future, the future hasn't happened, meaning there is no way I can know for sure the 100% true value of the outcome. Therefor I need a function that can minimize the error for my predicted future outcome, so the prediction can be as close as possible to the 100% true value. The frame work for solving this issue is we split the data I have into two sections, Training data: 75%-80% of the data, and Testing data: 20%-25%. We run our regression on the training sample and then we keep a portion of the data for testing purposes. "Initially, all training data is grouped in the same partition. The algorithm then begins allocating the data into the first two partitions or branches, using every possible binary split. The algorithm uses the split that minimizes the sum of squared deviations (varience of the set of values) and allocates the data into two more seperate partitions. The process continues until each node reaches a minimum node size and becomes a terminal node (Regression values at end of tree)" (FrontlineSolvers staff, 2022).
